# Image modes

# Converting a color image to Gray Scale

```python
im_gray= im.convert("L")
print(im_gray.mode)
```

```python
im_gray
```

# Converting a Gray image to Binary

```python
threshold = 100
img_bin = im_gray.point(lambda x : 255 if x> threshold else 0)
print(img_bin.mode)
#threshold is inversely prop to white cells
```

```python
img_bin
```

# Display Results as in Grid Form

#### The subplot() function takes three arguments that describe layout of figure .The Layout is organised in rows and columns , which are represented by the first and second argument, plt.subplot(1,2,1): It means the figure has 1 row , 2 columns and this plot is first plot.

```python
# Display results
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plt.subplot(2,3,1)
plt.imshow(im)
plt.axis('off')
plt.title('Original',fontsize=18)

plt.subplot(2,3,2)
plt.imshow(im_gray, cmap="prism")
plt.axis('off')
plt.title('gray',fontsize=18)

plt.subplot(2,3,3)
plt.imshow(img_bin,cmap="gray")
plt.axis('off')
plt.title('Binary',fontsize=18)

plt.show()
```

# what happens if we want to convert colors to binary direct

```python
im1= Image.open("lena.jpg")
imglena = Image.open("lena.jpg")
red_channel,blue_channel,green_channel = im1.split()
# cmaps only for color schemes, img.split() for split org img to r,g,b tuples stored in format & intensity values decided by split
plt.figure(figsize=(10,8))
plt.subplot(2,4,1)
plt.imshow(im1)
plt.axis('off')
plt.title('original',fontsize=18)

plt.subplot(2,4,2)
plt.imshow(red_channel,cmap='Reds')
plt.axis('off')
plt.title('Red',fontsize=18)

plt.subplot(2,4,3)
plt.imshow(blue_channel,cmap='Blues')
plt.axis('off')
plt.title('Blue',fontsize=18)

plt.subplot(2,4,4)
plt.imshow(green_channel,cmap='Greens')
plt.axis('off')
plt.title('Green',fontsize=18)

plt.show()
```
