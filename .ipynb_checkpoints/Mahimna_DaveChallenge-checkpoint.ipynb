# 1.Add sub multiply

```python
import numpy as np
from PIL import Image,ImageChops


imag1=Image.open(r"imag1.jpeg")
imag2=Image.open(r"imag2.jpeg")
#applying add method
imag3=ImageChops.add(imag1,imag2,scale=1.0,offset=2)
print("Image 1&2 Addition")
imag3
```

```python
imag4=ImageChops.subtract(imag1,imag2)
print("Image 1&2 Subtraction")
imag4
```

```python
imag5=ImageChops.multiply(imag1,imag2)
print("Image 1&2 Multiplied ")
imag5
```

# 3.Decrease contrast and enhancing image

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image
image_path = "imag1.jpeg"
image = Image.open(image_path)

# Convert image to graySCALE if its a color image
if image.mode != "L":
    image=image.convert('L')

# Compute the histogram
histogram = image.histogram()

# Plot histogram in blue color
plt.figure()
plt.title("Histogram of Pixel Intensities")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.plot(histogram,color='blue')
plt.show()
image
```

# Erode and dialate image using disk size 5

```python
from skimage.io import imread , imshow
from skimage import morphology
from PIL import Image,ImageFont,ImageDraw
```

```python
disk=morphology.disk(5)
print(disk)
```

```python
red,green,blue=imag1.split()

redmor=morphology.dilation(red,disk)
greenmor=morphology.dilation(green,disk)
bluemor=morphology.dilation(blue,disk)
dilr=Image.fromarray(redmor)
dilg=Image.fromarray(greenmor)
dilb=Image.fromarray(bluemor)
channels=(dilr,dilg,dilb)
rgb_image= Image.merge('RGB',channels)
rgb_image

```

```python
redmorp=morphology.erosion(red,disk)
greenmorp=morphology.erosion(green,disk)
bluemorp=morphology.erosion(blue,disk)
eror=Image.fromarray(redmorp)
erog=Image.fromarray(greenmorp)
erob=Image.fromarray(bluemorp)
channelserode=(eror,erog,erob)
rgbe_image= Image.merge('RGB',channelserode)
rgbe_image
```
